```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps
```

```python
file_list = ['A','B','C','D','E']

for i in file_list:
    
    file_path = 'augmentation_image\{}_cl\\'.format(i)
    file_names = os.listdir(file_path)
    total_origin_image_num = len(file_names)
    augment_cnt = 1
    num_augmented_images_A = 300
    num_augmented_images_B = 200
    num_augmented_images_C = 20
    num_augmented_images_D = 300
    num_augmented_images_E = 200
    
    if i == 'A':
    
        for j in range(1,num_augmented_images_A):
            change_picture_index = random.randrange(1, total_origin_image_num-1)
            print(change_picture_index)
            print(file_names[change_picture_index])
            file_name = file_names[change_picture_index]
    
            origin_image_path = file_path + file_name
            print(origin_image_path)
            image = Image.open(origin_image_path)
            random_augment = random.randrange(1,3)
    
            if(random_augment == 1):
            #이미지 좌우 반전
                print("invert")
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
            elif(random_augment == 2):
            #이미지 기울이기
                print("rotate")
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
            augment_cnt += 1
    
    elif i == 'B':
        
        for j in range(1,num_augmented_images_B):
            change_picture_index = random.randrange(1, total_origin_image_num-1)
            print(change_picture_index)
            print(file_names[change_picture_index])
            file_name = file_names[change_picture_index]
    
            origin_image_path = file_path + file_name
            print(origin_image_path)
            image = Image.open(origin_image_path)
            random_augment = random.randrange(1,3)
    
            if(random_augment == 1):
            #이미지 좌우 반전
                print("invert")
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
            elif(random_augment == 2):
            #이미지 기울이기
                print("rotate")
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
            augment_cnt += 1
    
    elif i == 'C':
        
        for j in range(1,num_augmented_images_C):
            change_picture_index = random.randrange(1, total_origin_image_num-1)
            print(change_picture_index)
            print(file_names[change_picture_index])
            file_name = file_names[change_picture_index]
    
            origin_image_path = file_path + file_name
            print(origin_image_path)
            image = Image.open(origin_image_path)
            random_augment = random.randrange(1,3)
    
            if(random_augment == 1):
            #이미지 좌우 반전
                print("invert")
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
            elif(random_augment == 2):
            #이미지 기울이기
                print("rotate")
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
            augment_cnt += 1
            
            
    elif i == 'D':
        
        for j in range(1,num_augmented_images_D):
            change_picture_index = random.randrange(1, total_origin_image_num-1)
            print(change_picture_index)
            print(file_names[change_picture_index])
            file_name = file_names[change_picture_index]
    
            origin_image_path = file_path + file_name
            print(origin_image_path)
            image = Image.open(origin_image_path)
            random_augment = random.randrange(1,3)
    
            if(random_augment == 1):
            #이미지 좌우 반전
                print("invert")
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
            elif(random_augment == 2):
            #이미지 기울이기
                print("rotate")
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
            augment_cnt += 1
    
    
    elif i == 'E':
        
        for j in range(1,num_augmented_images_E):
            change_picture_index = random.randrange(1, total_origin_image_num-1)
            print(change_picture_index)
            print(file_names[change_picture_index])
            file_name = file_names[change_picture_index]
    
            origin_image_path = file_path + file_name
            print(origin_image_path)
            image = Image.open(origin_image_path)
            random_augment = random.randrange(1,3)
    
            if(random_augment == 1):
            #이미지 좌우 반전
                print("invert")
                inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
                inverted_image.save(file_path + 'inverted_' + str(augment_cnt) + '.png')
        
            elif(random_augment == 2):
            #이미지 기울이기
                print("rotate")
                rotated_image = image.rotate(random.randrange(-20, 20))
                rotated_image.save(file_path + 'rotated_' + str(augment_cnt) + '.png')
        
            augment_cnt += 1
    
```
